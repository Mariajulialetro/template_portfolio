# **Panorama Covid-19 no Brasil**

COVID-19 é uma doença infecciosa causada por recém-descoberto coronavírus.

Transmitido principalmente por meio de gotículas provenientes de tosses ou espirros de pessoas infectadas, a gravidade dos sintomas varia muito de pessoa para pessoa.
Fato é, não se sabe muita coisa a respeito do COVID-19. Estudos estão sendo realizados no mundo todo, porém os resultados ainda não são conclusivos e definitivos.

familia-ilustrada-protegida-do-virus_52683-38847.jpg

Até o presente momento, observa-se que cerca de 80% dos casos confirmados são assintomáticos e rápidos. A maioria das pessoas que se encaixam nesse grupo, se recupera sem nenhuma sequela.

No entanto, 15% das pessoas terão infecções graves e precisarão de oxigênio. O restante das pessoas, que representam 5%, serão classificadas como infecções muito graves e precisarão de ventilação assistida, por meio de respiradores mecânicos em ambiente hospitalar.

Com o objetivo de elevar a consciência situacional a respeito do COVID-19 no Brasil, irei realizar uma análise sobre os dados públicos da doença.

# **Análise exploratória dos Dados**

Abaixo será importado os pacotes necessários para dar inicio a análise exploratória.


```python
#Importar as bibliotecas necessárias

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style()
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')

#Importção do dataset Covid-19 para o dataframe

df = pd.read_csv('https://raw.githubusercontent.com/carlosfab/dsnp2/master/datasets/owid-covid-data.csv')
```

```python
# Verificação da dimensão do dataset

print ('Dimensões do Dataset', '\nVariáveis:', df.shape[1], '\nentradas', df.shape[0])

#Verificando as primeiras entradas

df.head()
```

```python
#Extrair o nome das colunas (Series)

df.columns
```

```python
#Verificação do tipo das variáveis (Se são numeros inteiros ou do tipo float e strings)

df.dtypes
```

```python
#Verificação dos dados ausentes

round((df.isnull().sum()/df.shape[0]). sort_values(ascending=False),2)
```

```python
#Transformar a coluna data em Datatime

df. date= pd.to_datetime(df.date)
```

```python
#Verificando a data mais recente do dataset (Com essa informação, será possivel identificar o país lider de casos daquele momento)

df.date.max()
```

```python
#Total de casos

df.loc[df.date== '2020-06-15'].sort_values(by='total_cases',ascending=False)
```

Verificado na tabela acima os Paises lideres em casos em 15/06/2020 eram



*   Estados Unidos - 2094069.0
*   Brasil - 867624.0
*   Russia - 528964.0

Agora será verificado os casos de morte.






```python
# Total de mortes
df.loc[df.date=='2020-06-15'].sort_values(by='total_deaths', ascending= False)
```

Em relação as mortes, os locais que lideravam em 15/06/2020 eram:



*   Estados Unidos - 115732.0
*   Brasil - 43332.0
*   Reino Unido - 41698.0

Para melhor visualização será plotado um gráfico de barras abaixo:









```python
#Grafico de barras para total de mortes
top_5_total_deaths = df.loc[df.date=='2020-06-15', ['location', 'total_deaths']].sort_values(by="total_deaths", ascending=False)[1:6]

fig, ax = plt.subplots(figsize=(10,6))

ax.bar('location', 'total_deaths', data= top_5_total_deaths)

plt.tight_layout()
```

#**Como o Vírus está progredindo?**

Uma das melhores formas de visualizarmos a evolução ao longo do tempo, é o gráfico de linhas. Por isso, vamos plotar um gráfico de linha para demonstrar a evolução de mortes no mundo

```python
#Gráfico de linhas para mortes no mundo

df.loc[df.location == 'World', 'total_deaths'].reset_index(drop=True).plot(figsize=(8,6))
```

#**Análise Exploratória para o Brasil**

Agora que já temos uma noção dos acontecimentos no mundo, vamos focar nos acontecimentos no Brasil.

```python
#Criar uma cópia do DataFrame apenas com o Brasil
df_brasil = df.loc[df.location =='Brazil'].copy()

df_brasil.head()
```

```python
df_brasil.date.max()
```

Acima, verificamos que nosso dataset inclui datas desde 31-12-2019 até 15-06-2020.

Vamos verificar quando foram identificados o primeiro caso e a primeira morte no Brasil.

```python
# Identificar quando foi registrada a primeira morte no Brasil
df_brasil.loc[df_brasil['new_deaths']!=0.0].head(1)
```

```python
print('A primeira morte foi registrada em {}'. format(df_brasil.date.loc[3175]))
```

Plotar gráfico de linhas para verificar a evolução de casos e mortes no Brasil


```python
#Plotar gráficos para casos e mortes totais

x = df_brasil['total_cases']
y = df_brasil['total_deaths']

fig,(ax1, ax2)= plt.subplots(nrows=1, ncols=2, figsize=(9,7))

ax1.set_title('Casos no Brasil')
ax1.set_xlabel('Dias')
ax1.set_ylabel('Casos')
ax1.plot(x)

ax2.set_title('Total Mortes no Brasil')
ax2.set_xlabel('Dias')
ax2.set_ylabel('Mortes')
ax2.plot(y)

```
